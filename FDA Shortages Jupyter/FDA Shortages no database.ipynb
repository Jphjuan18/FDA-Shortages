```python
status_l = []
```

```python
count = 0
while count < len(divs):
    for i in divs[count].find_all('strong'):
        status_l.append(i)
        count += 1
       # print('------------')
        #print(i)
        #print(test)
        #print('-----------')
        #print(type(i))
```

```python
status_f = []
```

```python
for i in status_l:
    w = i.string.strip()
    #print(i)
    status_f.append(w)
```

```python
href_l = []
title_l = []
```

```python
#filter out the second 'a' which included the information needed
count = 0
subcount = 0
while count < len(divs):
    for i in divs[count].find_all('a'):
            i.find_all('a')
            href = i.get('href')
            href_l.append(href)
            #print('href: ')
            #print(href)
            #print(count) 
            title = i.get('title')
            title_l.append(title)
            #print('title: ')
            #print(title)
            #print(count) 
            #print('\n')
            subcount +=1
    if subcount == 2:
            subcount = 0
            count += 1
```

```python
href_f = []
```

```python
#filter href_l
for i in range(1,len(divs)*2,2):
    words = href_l[i].split(' ')
    wordsjoin = '%20'.join(words)
    href_f.append(wordsjoin)
    #print(wordsjoin)
    #print(i)
 
```

```python
S_URL = 'https://www.accessdata.fda.gov/scripts/drugshortages/'
```

```python
href_link = []
```

```python
for i in href_f: 
    href_link.append(S_URL + i)
```

```python
###href_link
```

```python
title_f = []
```

```python
#filter title_l
for i in range(1,len(divs)*2,2):
    q = title_l[i]
    w = q.replace(' detail', '')
    title_f.append(re.sub('Link to ', '', w))   
```

```python
###title_f
```

```python
print(len(title_f))
print(len(href_link))
print(len(status_f))
```

```python
subs = {'Drug': [], 'Link':[],'Status':[]}
```

```python
subs['Drug'] = title_f
subs['Link'] = href_link
subs['Status'] = status_f
```

```python
#Create a new dataframe
sub_df = pd.DataFrame(subs, columns=['Drug','Link','Status'])
```
